```python
import pandas as pd

nodes = pd.read_csv(
    '../data/nodes.csv'
)
nodes
```

```python
links = pd.read_csv(
    '../data/links_2023.csv'
)
links
```

```python
# filter nodes by degree (i.e. if a node has no links it has to go)

linked_nodes = set()
for _, link in links.iterrows():
    linked_nodes.add(link.src_id)
    linked_nodes.add(link.dst_id)
    
filtered_nodes = nodes.loc[nodes.ascii_id_company.isin(linked_nodes), :]
filtered_nodes
```

```python
import networkx as nx

# build graph
g = nx.Graph()

# add nodes
node_container = []
for _, company in filtered_nodes.iterrows():
    company = company.to_dict()
    node_id = company.pop('ascii_id_company')
    node_container.append((node_id, company))

g.add_nodes_from(node_container)

# add links
for _, link in links.iterrows():
    g.add_edge(link.src_id, link.dst_id, count = link.count)
```

```python
nx.draw(g, node_size = [n[1] for n in g.degree()])
```
