```python
import networkx as nx

# build graph
g = nx.DiGraph()

# add nodes
node_container = []
for _, company in nodes.iterrows():
    company = company.to_dict()
    node_id = company.pop('ascii_id_company')
    node_container.append((node_id, company))

g.add_nodes_from(node_container)

# add links
for _, link in links.iterrows():
    g.add_edge(link['src_id'], link['dst_id'], count = link['count'])
```

```python
ccs = list(nx.components.connected_components(g.to_undirected()))
lcc = g.subgraph(ccs[0])
pos = nx.drawing.layout.spring_layout(lcc)
```

```python
def find_max(tuples):
    max_value = 0
    max_value_company = None
    for t in tuples:
        if t[-1] > max_value:
            max_value = t[-1]
            max_value_company = t[0]
    
    return max_value, max_value_company
            

lcc.nodes()[find_max(lcc.degree())[-1]]
```

```python
lcc_edges = []
for edge in lcc.edges(data = True):
    lcc_edges.append([edge[0], edge[1], edge[2]['count']])

lcc_edges = pd.DataFrame(lcc_edges, columns = ['src_id', 'dst_id', 'count'])
lcc_edges.to_csv(
    '../data/lcc_edges.csv',
    index = False
)
```

```python
from collections import defaultdict

node_list = lcc.nodes(data = True)

intra = defaultdict(lambda: 0)
inter = defaultdict(lambda: 0)
for edge in lcc.edges(data = True):
    src, dst = edge[:-1]
    country_src = node_list[src]['country']
    country_dst = node_list[dst]['country']
    if country_dst == country_src:
        intra[src] += 1
        intra[dst] += 1
        
    else:
        inter[src] += 1
        inter[dst] += 1

intra = pd.Series(intra)
intra.name = 'intra_edges'
inter = pd.Series(inter)
inter.name = 'inter_edges'

intra_vs_inter = pd.concat([intra, inter], axis = 1).reset_index(names = ['ascii_id_company'])
intra_vs_inter
nodes.merge(
    intra_vs_inter,
    on = 'ascii_id_company',
    how = 'left'
).to_csv(
    '../data/nodes_with_some_region_and_intra_vs_inter_count.csv',
    index = False
)
```

```python
groups = {}

for node_id, node_attrs in lcc.nodes(data = True):
    country = node_attrs['country']
    if not country in groups:
        groups[country] = set([node_id])
        
    else:
        groups[country].add(node_id)

node_lists, list_names = [], []
for name, node_list in groups.items():
    node_lists.append(node_list)
    list_names.append(name)
    
gbc = nx.group_betweenness_centrality(lcc.to_undirected(), node_lists)
```

```python
gbc_df = pd.DataFrame(
    {'group_betweeness_centrality': gbc},
    index = list_names
)
gbc_df
```

```python
global_average_shortest_path_length = nx.average_shortest_path_length(lcc.to_undirected())
all_nodes = set(lcc_nodes.ascii_id_company)

results = {
    'd_aspl': {},
    'd_cc': {},
    'disconnections': {},
}
companies_per_country = lcc_nodes.groupby('country').count()['ascii_id_company']
disconnect_percentages = []
for country, companies in lcc_nodes.groupby('country'):
    country_nodes = set(companies.ascii_id_company)
    subgraph = lcc.subgraph(all_nodes - country_nodes)
    ncc = len(list(nx.components.connected_components(subgraph.to_undirected())))
    
    n_zero_degree, zero_degree_nodes = 0, []
    for n, d in subgraph.degree():
        if not d:
            n_zero_degree += 1
            zero_degree_nodes.append(n)

    sublcc = lcc_nodes.loc[lcc_nodes.ascii_id_company.isin(zero_degree_nodes), :]
    disconnects_per_country = sublcc.groupby('country').count()['ascii_id_company']
    disconnect_percentage = disconnects_per_country / companies_per_country * 100
    disconnect_percentage.name = country
    disconnect_percentages.append(disconnect_percentage) 
    
    if ncc == 1:
        aspl = nx.average_shortest_path_length(subgraph.to_undirected())
    
    else:
        aspl = np.inf
        
    results['d_aspl'][country] = aspl - global_average_shortest_path_length
    results['d_cc'][country] = ncc
    results['disconnections'][country] = n_zero_degree
```

```python
def to_processing_array(df, grouping_column, filename):
    template = 'blocks: 3, gaps: [{g1}, {g2}, {g3}], colors: [{c1}, {c2}, {c3}], orientation: {o}'
    orientation = 0
    next_orientation = 1
    color_ids = {g: i + 1 for i, g in enumerate(df[grouping_column].unique())}
    
    with open(filename, 'w') as arrayfile: 
        for g, i in color_ids.items():
            arrayfile.write(f'{g} = {i}\n')
        
        gs, cs = {}, {}
        count = 1
        for i, (_, country) in enumerate(df.iterrows()):
            gs[f'g{count}'] = np.random.uniform(0, 5)
            cs[f'c{count}'] = color_ids[country.region]
            count += 1
            
            if not i:
                continue
            
            if not i%3:
                output = template.format(
                    **gs,
                    **cs,
                    o = orientation
                )
                arrayfile.write('{ ' + output + ' },\n')
                tmp = orientation
                orientation = next_orientation
                next_orientation = tmp
                count = 0
                
                
to_processing_array(
    df, 
    'region', 
    '../results/testarray.txt'
)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools as it

def make_random_scatter_grid(num_nodes, vert_lines, horiz_lines, scaling = 2):
    x_positions, y_positions = [], []
    max_index = vert_lines * horiz_lines
    rand_index = set(
        np.random.choice(
            np.arange(max_index), 
            replace = False,
            size = num_nodes
        )
    )
    
    x_positions, y_position = [], []
    for i, (xpos, ypos) in enumerate(it.product(range(vert_lines), range(horiz_lines))):
        if i in rand_index:
            x_positions.append(xpos * scaling)
            y_positions.append(ypos * scaling)
            
            
    return x_positions, y_positions


def compute_edge_coordinates(grid, edge_list):
    edges = []
    for nodes in edge_list:
        nodes_pos = grid.loc[nodes, :]
        xmin_idx = nodes_pos.xpos.idxmin()
        xmax_idx = nodes_pos.xpos.idxmax()
        ymin_idx = nodes_pos.ypos.idxmin()
        ymax_idx = nodes_pos.ypos.idxmax()
        direction = np.random.randint(2)
        xmax_below_xmin = nodes_pos.ypos[xmin_idx] > nodes_pos.ypos[xmax_idx]
        
        if direction: # up __|
            x_idx = xmax_idx
            y_idx = ymax_idx if xmax_below_xmin else ymin_idx
            edge_corner_xpos = int(nodes_pos.xpos[x_idx])
            edge_corner_ypos = int(nodes_pos.ypos[y_idx])
        
        else: # down |--
            x_idx = xmin_idx
            y_idx = ymin_idx if xmax_below_xmin else ymax_idx
            edge_corner_xpos = int(nodes_pos.xpos[x_idx])
            edge_corner_ypos = int(nodes_pos.ypos[y_idx])
        
        edges.append(
            [
                [int(nodes_pos.xpos[xmin_idx]), edge_corner_xpos, int(nodes_pos.xpos[xmax_idx])],
                [int(nodes_pos.ypos[xmin_idx]), edge_corner_ypos, int(nodes_pos.ypos[xmax_idx])]
            ]
        )
        
    return edges


def induce_subgraph(g, node_attr_name, node_attr_value):
    considered_nodes = []
    for node_id, node_attrs in g.nodes(data = True):
        if node_attrs[node_attr_name] != node_attr_value:
            continue
            
        considered_nodes.append(node_id)
    
    return g.subgraph(considered_nodes)


def plot_edges(edge_coordinates, ax):
    colors = ['#30303a']
    rng = np.random.uniform(0, 1, size = len(colors))
    probabilities = rng/rng.sum()
    color_selection = np.random.choice(colors, size = len(edge_coordinates), p = probabilities)
    for c, (edge_x, edge_y) in zip(color_selection, edge_coordinates):
        ax.plot(edge_x, edge_y, c = c, zorder = 1)


def subgraph_to_random_grid(g):
    num_nodes = g.number_of_nodes()
    num_grid_points = num_nodes * 75
    num_square_lines = int(np.ceil(np.sqrt(num_grid_points)))
    vert_lines = int(np.ceil(num_square_lines * 16 / 9))
    horiz_lines = num_square_lines - vert_lines + num_square_lines
    xpos, ypos = make_random_scatter_grid(
        num_nodes,
        vert_lines,
        horiz_lines
    )
    
    data = {}
    for x, y, node in zip(xpos, ypos, g.nodes()):
        data[node] = (x, y)
        
    df = pd.DataFrame.from_dict(data, orient = 'index', columns = ['xpos', 'ypos'])
    return df


def draw_random_grid_graph(g, ax, markers = 'o'):
    grid = subgraph_to_random_grid(g)
    edges = compute_edge_coordinates(grid, g.edges())
    plot_edges(edges, ax)
    sizes = np.array([(np.log(d) + 0.15) * 100 for n, d in g.degree()])
    marker_assignments = np.random.choice(list(markers), size = len(grid))
    colors = ['#fac901', '#225095', '#dd0100', '#30303a']
    rng = np.random.uniform(0, 1, size = len(colors))
    probabilities = rng/rng.sum()
    color_selection = np.random.choice(colors, size = len(grid), p = probabilities)
    for x, y, m, s, c in zip(grid.xpos, grid.ypos, marker_assignments, sizes, color_selection):
        ax.scatter(
            x, y, 
            marker = m, 
            s = s, 
            facecolor = 'none', 
            edgecolor = c, 
            linewidth = 1.5
        )
        
    # for i, m in enumerate(markers):
    #     ax.scatter(
    #         grid.xpos[marker_assignments == i], 
    #         grid.ypos[marker_assignments == i], 
    #         zorder = 2, 
    #         s = sizes[marker_assignments == i],
    #         marker = m,
    #         facecolors = 'none',
    #         edgecolors = '#30303a',
    #         linewidths = 1.5
    #     )
    
    # cosmetics
    ax.set_xticks([])
    ax.set_yticks([])
    for pos in ['left', 'right', 'top', 'bottom']:
        ax.spines[pos].set_visible(False)
    
subgraph = induce_subgraph(lcc, 'country', 'Japan')

for i in range(25):
    fig, ax = plt.subplots()
    draw_random_grid_graph(subgraph, ax, 's')
    fig.set_figwidth(7.5)
    fig.set_figheight(2.5)
    fig.tight_layout()
    fig.savefig(f'../plots/japan_micro_{i}_modrian.pdf')
```
